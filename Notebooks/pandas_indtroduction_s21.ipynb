```python
dict_values_replace = {
    'NEAR BAY':0, 
    '<1H OCEAN':10, 
    'INLAND':100, 
    'NEAR OCEAN':5, 
    'ISLAND':-1
}
 
```

```python
housing_df.loc[:,"replace_method_1"] = housing_df.ocean_proximity.apply(lambda x: dict_values_replace[x])
```

```python
housing_df
```

```python
housing_df["replace_method_2"] = housing_df.ocean_proximity.replace(dict_values_replace)
```

```python
housing_df
```

```python
for k in dict_values_replace.keys():
    print(f" for the key :{k}")
    print(f" will be replaced by the value {dict_values_replace[k]}")
    housing_df.loc[housing_df.ocean_proximity == k,"replace_method_3"] = int(dict_values_replace[k])
```

```python
housing_df
```

```python
(housing_df.replace_method_1 == housing_df.replace_method_2).value_counts()
```

```python
(housing_df.replace_method_1 == housing_df.replace_method_3).value_counts()
```
